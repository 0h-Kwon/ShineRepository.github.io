### 1. 데이터 준비

```
import pandas as pd
import tensorflow as tf
```

```
boston = pd.read_csv('https://raw.githubusercontent.com/hwaneGithub/hwaneGithub.github.io/master/Database/boston.csv')
boston.head(3)
```

```
x_data = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm',
                 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
y_data = boston[['medv']]
print(x_data.shape)
print(y_data.shape)
```

### 2. 모델 구조 생성

```
X = tf.keras.layers.Input(shape=[13]) # 입력층 구성(13개) == y = w1x1 + w2x2 + ... + w13x13 + b
H = tf.keras.layers.Dense(10, activation='swish')(X)
Y = tf.keras.layers.Dense(1)(H)       # 출력층 구성(1개)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
model.fit(x_data, y_data, epochs=10000, verbose=0)
model.fit(x_data, y_data, epochs=10)
```

```
model.summary()   # 모델이 잘 만들어 졌는지 확인
                  # Param : 학습하는 가중치의 개수
```

```
print("Predictions: ", model.predict(x_data[0:5]))    # 예측값
print(y_data[0:5])   # 실제값
```
