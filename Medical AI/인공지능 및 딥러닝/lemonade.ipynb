# 레모네이드 판매 예측 모델

### 1. 데이터 준비

```
import pandas as pd
```

```
lemonade = pd.read_csv('https://raw.githubusercontent.com/hwaneGithub/hwaneGithub.github.io/master/Database/lemonade.csv')
lemonade.head(3)
```

```
x_data = lemonade[['온도']]       # 독립변수 : 원인
y_data = lemonade[['판매량']]     # 종속변수 : 결과
print(x_data.shape, y_data.shape)
```

### 2. 모델 구조

```
import tensorflow as tf
```

```
X = tf.keras.layers.Input(shape=[1])  # 독립변수 컬럼 수
Y = tf.keras.layers.Dense(1)(X)       # 종속변수 컬럼 수
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')             # model이 학습할 방법
```

### 3. 모델 학습

```
model.fit(x_data, y_data, epochs=10000, verbose=0)  # epochs: 전체 데이터를 몇 번 학습할지
model.fit(x_data, y_data, epochs=10)                # verbose=0 : 학습이 출력되지 않음
                                                    # loss : (실제값 - 예측값)**2의 평균.
                                                    # 즉, 실제와 예측의 차이가 적을수록 loss가 0에 가까워짐.
                                                    # e-04 = 10**-4 =>  2.0868e-04 = 2.0868 * 0.0001 = 0.0002
```

4. 모델 이용

```
model.predict(x_data)
```

```
print('예측값 : ', model.predict([[15]]))   # 온도가 15도일 때 몇 개 사용하면 되는지
```
